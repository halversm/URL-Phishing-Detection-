```python
import pandas as pd

# Import dataset
dataset = pd.read_csv('dataset.csv')
dataset.head()
```

```python
# Data exploration
dataset.info()
dataset.describe()
```

```python
# Check for missing values

dataset.isnull().sum()
```

```python
# Drop 'Index' column
dataset.drop('index', axis = 1, inplace = True)
```

```python
# Explore class distrubition, -1 = phishing and 1 = legitimate

dataset['Result'].value_counts()
```

```python
import numpy as np

# Change outcome of 'Results' column so 0 = legitimate, 1 = phishing 
dataset['Result'] = np.where(dataset['Result'] == -1, 1, 0)

# Drop statistical report
dataset.drop('Statistical_report', axis=1, inplace=True)
dataset.drop('SSLfinal_State', axis=1, inplace=True)
print(dataset.columns)
```

```python
# Explore class distribution again, 0 = legitimate, 1 = phishing
dataset['Result'].value_counts()
```

```python
# Heat Map of correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize =(8,8))
sns.heatmap(dataset.corr())


# Calculate correlations between the target variable and other variables
target_correlations = dataset.corr()["Result"].sort_values(ascending=False)

# Print the correlations
print("Correlation with the target variable (", "Result", "):")
print(target_correlations)
```

```python
# Visualize 'Result' column

import matplotlib.pyplot as plt

# Count the occurrences of each unique value
result_counts = dataset['Result'].value_counts()

# Plot bar chart
plt.bar(result_counts.index, result_counts.values, color='skyblue')
plt.xlabel('Result')
plt.ylabel('Count')
plt.title('Count of Results')
plt.show()

# Pie chart

# Count the occurrences of each unique value
result_counts = dataset['Result'].value_counts()

# Map numerical labels to descriptive categories
label_mapping = {0: 'Legitimate', 1: 'Phishing'}
result_counts.index = result_counts.index.map(label_mapping)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(result_counts, labels=result_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Results')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()



```

```python
# Split dataset

X = dataset.iloc[:, :-1]  # Exclude outcome 'Result'
y = dataset.iloc[:, -1]  # Include only target variable 'Result'
```

```python
# Splitting the dataset into the training set and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

# Initialize MLPClassifier
mlp_classifier = MLPClassifier(hidden_layer_sizes=(16, 16, 8), activation='relu', solver='adam', max_iter=500, random_state=42)

# Fit MLPClassifier to training set
mlp_classifier.fit(X_train, y_train)

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions on test data
y_pred = mlp_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)

# F1-score
f1 = f1_score(y_test, y_pred)
print("F1-Score:", recall)
```

```python
# *** Implement Cross Validation *** 
from sklearn.model_selection import cross_val_score

# Perform cross-validation
cv_scores = cross_val_score(mlp_classifier, X_train, y_train, cv=5)  

# Print the cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Mean accuracy:", cv_scores.mean())
```

```python
# *** Implement Random Forest ***

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_classifier = rf_model.fit(X_train, y_train)
```

```python
# Make predictions on test data

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = rf_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)

# F1-score
f1 = f1_score(y_test, y_pred)
print("F1-Score:", f1)
```

```python
# *** Implement Cross Validation *** 
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(rf_classifier, X_train, y_train, cv=5)  

# Print the cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Mean accuracy:", cv_scores.mean())
```
